```python

from functools import total_ordering
```

```python

class MontanaPeaks:

  #init function
  def __init__(self, peak_string,height):
    self.peak_string = peak_string
    self.height = height

  def __repr__(self) -> str:
      return 'Object: {}'.format(self.peak_string)

  def __add__(self, other):
    return self.peak_string +  ' ' + other


  def __len__(self):
    return self.height
```

```python

class Farm_Animals:

  def __init__(self,Name,animal_type) -> None:
      self.Name = Name
      self.animal_type = animal_type
      self.children = []
      self.daily_milk = []

  def __repr__(self) -> str:
      #return "test"
      return f'{self.__class__.__name__} ({self.Name}, {self.animal_type})'

  def __str__(self) -> str:
      return f'The class object is: {self.__class__.__name__} the init values are: {self.Name}, {self.animal_type}'

  def add_milk_production(self,ivolume):
      if not isinstance(ivolume,int):
        raise ValueError('Volume amount must be an integer value')
      
      self.daily_milk.append(ivolume)

  @property
  def how_much_milk(self):
    return sum(self.daily_milk)

  def __len__(self):
    return len(self.daily_milk)

  def __getitem__(self,ipos):
    return self.daily_milk[ipos]


    
```

```python

@total_ordering
class Milk_Customers:

  def __init__(self, Name, vehicle, address):
    self.Name = Name
    self.vehicle = vehicle
    self.address = address
    self.purchase_log = []

  def __repr__(self) -> str:
      return f'{self.__class__.__name__} ({self.Name}, {self.vehicle}, {self.address})'

  def get_milk(self, iGals):
    if not isinstance(iGals, int):
      raise ValueError('iGals value must be an integer value')

    self.purchase_log.append(iGals)

  @property
  def how_many_gals(self):
    return sum(self.purchase_log)

  def __eq__(self, other) -> bool:
      return self.how_many_gals == other.how_many_gals

  def __lt__(self,other):
    return self.how_many_gals < other.how_many_gals

  def __getitem__(self,ipos):
    self.purchase_log[ipos]

  def __add__(self,other):
    new_owner = f'{self.Name} & {other.Name}'
    new_vehicle = f'{self.vehicle} & {other.vehicle}'
    new_address = f'{self.address} & {other.address}'
    new_milk_customer = Milk_Customers(new_owner,new_vehicle,new_address)

    for new_purchase in list(self.purchase_log) + list(other.purchase_log):
      new_milk_customer.get_milk(new_purchase)

    return new_milk_customer
```

```python

cust1 = Milk_Customers('seth','diesel pickup', '155 smith drive')

cust1.get_milk(2)
cust1.get_milk(1)
cust1.get_milk(3)
cust1.get_milk(2)
cust1.get_milk(4)

cust2 = Milk_Customers('mike','nissian leaf', '15 smith drive')

cust2.get_milk(5)
cust2.get_milk(4)
cust2.get_milk(6)
cust2.get_milk(2)
cust2.get_milk(1)

print(cust2 > cust1)


cust_merge = cust2 + cust1

print(cust_merge)
```
