```python
%%file food_driver.py

from dataclasses import dataclass

@dataclass
class Food_Driver:
  #info about driver
  name: str
  contractor_id: int
  delivery_tot: int
  miles_driven: int
  hours_active: int
  total_wage: float

  delivery_dict: dict

  delivery_cnt: int = 0

  


  def Average_Wage_Hour(self) -> float:
    avg_wage = self.total_wage / self.hours_active
    return avg_wage

  def Get_Dollar_Per_Mile(self) -> float:
    dollar_mile_val = self.total_wage / self.miles_driven
    return dollar_mile_val

  def Average_Delivery_Amount(self) -> float:
    avg_delivery_val = self.total_wage / self.delivery_tot
    return avg_delivery_val

  
  def Add_Single_Delivery_Trip(self, single_delivery_dict) -> None:

    #check to see delivery passed is the correct type
    if( isinstance(single_delivery_dict, dict) ):
      pass
    else:
      raise TypeError('expected a dictionary valiable')
      
    delivery_str_id = f'D{self.delivery_cnt}'
    self.delivery_cnt += 1
    self.delivery_dict[delivery_str_id] = single_delivery_dict
    return True

  
```

```python
%%file test_food_driver.py

from pytest import approx
import pytest

from food_driver import Food_Driver

test_driver = Food_Driver(name='seth', contractor_id=435,hours_active=6,total_wage=85.65,delivery_tot=15,miles_driven=45, delivery_dict={})

def test_Average_Wage_Hour():
  assert isinstance(test_driver.Average_Wage_Hour(), float)
  

def test_Average_Wage_Hour_return_float():
  assert (approx(test_driver.Average_Wage_Hour()) == 14.275)

def test_Dollar_Per_Mile():
  assert ( approx(test_driver.Get_Dollar_Per_Mile()) == 1.903333)

def test_Dollar_Per_Mile_return_float():
  assert isinstance(test_driver.Get_Dollar_Per_Mile(), float)

def test_Average_Delivery_Amount():
  assert (approx(test_driver.Average_Delivery_Amount()) == 5.71)

def test_Average_Delivery_Amount_return_float():
  assert isinstance(test_driver.Average_Delivery_Amount(), float)

def test_Add_Single_Delivery_Trip():
  delivery_example = {'place':'SubWay', 'Miles':4.5, 'Payout':8.75}
  assert  test_driver.Add_Single_Delivery_Trip(delivery_example) == True 

def test_Add_Single_Delivery_Trip_pass_dict():
  delivery_example = ['subway',4.5,8.75]
  
  with pytest.raises( TypeError, match='expected a dictionary valiable'):
    test_driver.Add_Single_Delivery_Trip(delivery_example)



```
