```python
peoples = [{'first':'seth','last':'foster'},{'first':'bill','last':'smith'},{'first':'jim','last':'baker'}]

email_domaim = 'flyfishers.com'
```

```python

def build_email_address():
  i_how_many_emails = 0

  def email_together(name,email_base):
    nonlocal i_how_many_emails
    full_email = f'{name}_mycompany@{email_base}'
    i_how_many_emails += 1
    return full_email, i_how_many_emails
  
  return email_together

```

```python

compute_emails = build_email_address()

for single_person in peoples:
  print(single_person)
  new_email, icnt = compute_emails(single_person['first'],email_domaim)
  print(f'Here is the new email: {new_email}')
  print(f'This many have been created: {icnt}')
```
