```python

from abc import ABC, abstractmethod

class Plow_request:
  def __init__(self):
    self.job_id = 'AB326'
    self.job_status = 'pending'

  def set_status(self, status):
    self.job_status = status

class Plow_Template(ABC):

  @abstractmethod
  def Set_rig_availible(self):
    pass

  @abstractmethod
  def is_rig_availible(self):
    pass


class Plow_Truck(Plow_Template):
  def __init__(self):
    self.rig_type = 'truck'
    self.rig_id = None
    self.rig_availible = False

  def Set_rig_id(self):
    #simple number setting rig id
    self.rig_id = 'T231'

  def Set_rig_availible(self):
    if(self.rig_id != None):
      self.rig_availible = True

  def is_rig_availible(self) -> bool:
    return self.rig_availible

  
class Plow_Blower(Plow_Template):
  def __init__(self):
    self.rig_type = 'push_blower'
    self.rig_id = None
    self.rig_availible = False

  def Set_rig_id(self):
    #simple number setting rig id
    self.rig_id = 'B451'

  def Set_rig_availible(self):
    if(self.rig_id != None):
      self.rig_availible = True

  def is_rig_availible(self) -> bool:
    return self.rig_availible



class Plow_Dozer(Plow_Template):
  def __init__(self):
    self.rig_type = 'dozer'
    self.rig_id = None
    self.rig_availible = False

  def Set_rig_id(self):
    #simple number setting rig id
    self.rig_id = 'D756'

  def Set_rig_availible(self):
    if(self.rig_id != None):
      self.rig_availible = True

  def is_rig_availible(self) -> bool:
    return self.rig_availible


class Perform_Plowing:

  def __init__(self, plower: Plow_Template):
    self.plower = plower


  def plow_road(self, Plow_request):
    
    self.plower.Set_rig_id()
    self.plower.Set_rig_availible()

    if( self.plower.is_rig_availible()):
      print('The rig is availible for plowing')
      Plow_request.set_status('in_process')
    else:
      print('The rig is not availible for plowing')
```

```python

my_plower = Plow_Truck()
plowing_job = Plow_request()
get_plowing = Perform_Plowing(my_plower)

get_plowing.plow_road(plowing_job)


my_plower = Plow_Dozer()
plowing_job = Plow_request()
get_plowing = Perform_Plowing(my_plower)

get_plowing.plow_road(plowing_job)


my_plower = Plow_Blower()
plowing_job = Plow_request()
get_plowing = Perform_Plowing(my_plower)

get_plowing.plow_road(plowing_job)


```
