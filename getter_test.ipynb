```python

class milk_cust:

  def __init__(self,name,address,distance,state):
    self.name = name
    self.address = address
    self.distance = distance
    self.state = state

  @property
  def email(self):
    return f'{self.name}.{self.state}@gmail.com'

  @property
  def long_name(self):
    return f'{self.name}-{self.address}-{self.state}'

  @long_name.setter
  def long_name(self,l_name):
    name,address,state = l_name.split('-')
  

  def define_place(self):
    return f'{self.name} to {self.address}'

milk_cus1 = milk_cust('brett','248 smith drive',17,'mt')

milk_cus1.long_name = 'jim-sunny loop-mt'

print(milk_cus1.name)
print(milk_cus1.address)
print(milk_cus1.distance)
print(milk_cus1.email)
print(milk_cus1.long_name)
```
