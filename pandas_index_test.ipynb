```python

populations = [33871648, 37253956,
               18976457, 19378102,
               20851820, 25145561]


state = ['California','California','New York','New York','Texas','Texas']
year = [2000,2010,2000,2010,2000,2010]

state_ser = pd.Series(state)
year_ser = pd.Series(year)
pop_ser = pd.Series(populations)
```
