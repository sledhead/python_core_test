```python

from functools import partial, cmp_to_key

def BuildFullName(firstname,lastname):
  full_name = f'{firstname}_{lastname}'
  return full_name


lst_name = partial(BuildFullName,lastname='foster')

print(lst_name('seth'))
print(lst_name('mike'))
```

```python

#trying out a custom sorting code

def cmp_special(a,b):
  #will compare str length
  if( len(a) > len(b)):
    return 1
  
  elif( len(a) < len(b)):
    return -1

  else:
    return 0


sample_lst = ['bill', 'kalispell', 'sunny', 'snow']

new_order = sorted( sample_lst, key=cmp_to_key(cmp_special) )

print(new_order)
```

```python

long_word = max(sample_lst, key=cmp_to_key(cmp_special) )
print(long_word)

full_lst = [{'name': 'seth', 'state':'mt', 'zipcode': 59901}, {'name': 'bill', 'state':'ca', 'zipcode': 45128}, {'name': 'jeff', 'state':'ma', 'zipcode': 12895}, {'name': 'jim', 'state':'id', 'zipcode': 68401}]

def cmp_state(a,b):
  #sort by state
  if( a['state'] > b['state'] ):
    return 1

  elif( a['state'] < b['state'] ):
    return -1

  else:
    return 0

def cmp_name(a,b):
  #sort by first name
  if( a['name'] > b['name'] ):
    return 1

  elif( a['name'] < b['name'] ):
    return -1

  else:
    return 0

def cmp_zipcode(a,b):
  #sort by zipcode
  if( a['zipcode'] > b['zipcode'] ):
    return 1

  elif( a['zipcode'] < b['zipcode'] ):
    return -1

  else:
    return 0

state_ordered = sorted(full_lst, key=cmp_to_key( cmp_state ))

print(state_ordered )

name_ordered = sorted(full_lst, key=cmp_to_key(cmp_name) )

print(name_ordered)

zipcode_ordered = sorted(full_lst, key=cmp_to_key(cmp_zipcode) )

print(zipcode_ordered)
```
