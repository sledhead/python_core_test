```python
import pandas as pd
df = pd.read_excel('https://github.com/datagy/mediumdata/raw/master/sample_pivot.xlsx', parse_dates=['Date'])
print(df.head())
```

```python
df.info()
```

```python
df.groupby('Region')['Sales'].count()
```

```python
df.groupby('Type')['Sales'].count()
```

```python
df.groupby(['Region','Type']).agg({'Sales':['count','sum'],'Units':'sum'})
```

```python
df
```

```python
pivot = pd.pivot_table(data=df, index='Region')
pivot
```

```python
pivot = pd.pivot_table(data=df, index='Region', values='Sales')
pivot
```

```python
pivot = pd.pivot_table(data=df, index='Region', values='Sales', aggfunc=['sum','mean'])
pivot
```
