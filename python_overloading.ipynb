```python

def TellFunctionName(func):
  #show caller the name of function called
  
  def innerfunction():
    print(f'Here is the name of the function calling {func.__name__}')
    func()
  
  return innerfunction


@TellFunctionName
def CoolMan():
  for i_num in range(1,5):
    print(f'CoolMan function {i_num}')


@TellFunctionName
def SecondMan():
  for i_v in range(1,5):
    print(f'SecondMan function {i_v}')


CoolMan()
SecondMan()
```

```python

#Decorator for testing the correct value passed to the function
def check_value_decorator(dataType,message1,message2):
  def the_decorator(func):
    #show caller which function is calling to setup...
    print(message1)

    def wrapper(*args, **kwargs):
      #Now the function is calling...
      print(message2)
      if all([type(arg) == dataType for arg in args]):
        return func(*args, **kwargs)
      else:
        wrong_type = [[type(arg),arg] for arg in args if type(arg) != dataType]
        for single_guy in wrong_type:
          print(f'The following value was passed into the function: {single_guy[1]} The type {single_guy[0]} is wrong, only this type is allowed: {dataType} ')
      return "Invalid Input See above message"
    return wrapper
  return the_decorator


@check_value_decorator(str, "Decorator for 'stringJoin'", "stringJoin started ...")
def stringJoin(*args):
    st = ''
    for i in args:
        st += i
    return st

  
@check_value_decorator(int, "Decorator for 'SumTheNum'", "SumTheNum started ...")
def SumTheNum(*args):
  i_tot = 0
  for the_num in args:
    i_tot = i_tot + the_num

  return i_tot


print(stringJoin("I ", 'like ', 6, 'for', "geeks"))

print(SumTheNum(4,5,'me',7,'thesnow'))
```
