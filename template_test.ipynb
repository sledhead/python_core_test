```python
from string import Template
```

```python

customer_sales = [('bill',35.45),('seth',55.65),('mike',12.45),('mary',75.35)]

customer_total = Template('Here is $name total grocery bill: $total')

for single_c in customer_sales:
  print( customer_total.substitute( name = single_c[0], total=single_c[1]))
```

```python

new_customer_total = Template("Here is ${name}'s total grocery bill: $total")

for single_c in customer_sales:
  print( new_customer_total.substitute( name = single_c[0], total=single_c[1]))
```
