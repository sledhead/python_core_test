```python

@dataclass
class FarmAnimals(ABC):
  name: str
  id: int
  animal_class: str
  animal_product: str

  @abstractmethod
  def compute_farm_value(self) -> int:
    """return the amount of value produced by the animal"""


@dataclass
class CowAnimal(FarmAnimals):

  inum_calf: int = 0
  inum_calf_price: int = 300
  inum_milk_gallons: int = 0
  inum_milk_gal_price: int = 4

  def compute_farm_value(self) -> int:
      return (self.inum_calf * self.inum_calf_price + self.inum_milk_gallons
               * self.inum_milk_gal_price)
      

@dataclass
class ChickenAnimal(FarmAnimals):

  inum_eggs: int = 0
  iprice_per_doz: int = 4

  def compute_farm_value(self) -> int:
      return (self.inum_eggs * self.iprice_per_doz)




```

```python

cow1 = CowAnimal('annie',34,'cow','milk',4,250,25,5)
print(cow1.compute_farm_value())

chicken1 = ChickenAnimal('sim',32,'chicken','egg',78,6)

print(chicken1.compute_farm_value())
```
