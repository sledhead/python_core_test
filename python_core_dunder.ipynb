```python

class MontanaPeaks:

  #init function
  def __init__(self, peak_string,height):
    self.peak_string = peak_string
    self.height = height

  def __repr__(self) -> str:
      return 'Object: {}'.format(self.peak_string)

  def __add__(self, other):
    return self.peak_string +  ' ' + other


  def __len__(self):
    return self.height
```

```python

class Farm_Animals:

  def __init__(self,Name,animal_type) -> None:
      self.Name = Name
      self.animal_type = animal_type
      self.children = []
      self.daily_milk = []

  def __repr__(self) -> str:
      #return "test"
      return f'{self.__class__.__name__} ({self.Name}, {self.animal_type})'

  def __str__(self) -> str:
      return f'The class object is: {self.__class__.__name__} the init values are: {self.Name}, {self.animal_type}'

  def add_milk_production(self,ivolume):
      if not isinstance(ivolume,int):
        raise ValueError('Volume amount must be an integer value')
      
      self.daily_milk.append(ivolume)

  @property
  def how_much_milk(self):
    return sum(self.daily_milk)

  def __len__(self):
    return len(self.daily_milk)

  def __getitem__(self,ipos):
    return self.daily_milk[ipos]


    
```
