```python

from dataclasses import dataclass
from datetime import datetime
```

```python

@dataclass
class email_requests:
  date: datetime
  from_name: str
  from_address: str
  request: str


@dataclass
class email_raw_reqs:
  date: str
  from_address: str
  request: str

```

```python

time_data = "25/05/99 02:35:5.523"
format_data = "%d/%m/%y %H:%M:%S.%f"
date_req = datetime.strptime(time_data, format_data)

e1 = email_requests(date_req,'seth','seth@email.com','status')
e2 = email_requests(date_req,'mike','mike@email.com','status')

print(e1)
print(e2)
```

```python

from attr import attrs, attrib, define
```

```python

@attrs
class simple_cl:
  from_name = attrib()
  request_topic = attrib()
  index_num = attrib(init=False,eq=False)
  level = attrib(default='1',eq=False, converter=int)
  temp_history = attrib(factory=list)
  alarm_val = attrib(default=25)

  @alarm_val.validator
  def check_alarm_val(self,attribute,value):
    #print(type(value))
    if( (value <= 24) or (value >= 125) ):
      raise ValueError(f'The alarm value provided is outside of the required range: Must be between 25 and 125')


  
@define
class simple_test:
  name: str
  age: int
  address: str
  state: str





  
```

```python

gg = simple_cl('martin','status')
tt = simple_cl('martin','status')

kk = simple_cl('martin','status',level=4,alarm_val=26)

vv = simple_cl('mike','status',level='3')



print(gg == tt)

print(gg)

print(id(gg.temp_history))
print(id(tt.temp_history))

print(kk)

print(vv)

pp = simple_test(name='pat',age=25,address='345 smith dirve',state='mt')

print(pp)
```
