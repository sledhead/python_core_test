```python
df_period = df.copy()
```

```python
time_dict = {1700:{'begin':1700, 'end':1749}, 1750:{'begin':1750, 'end':1799}, 1800:{'begin':1800, 'end':1849},
              1850:{'begin':1850, 'end':1899}, 1900:{'begin':1900, 'end':1949}, 1950:{'begin':1950, 'end':1999} }

def get_time_section(date):
  rtr_key = 1200
  range_list = []
  index_s = []
  for items in date.iteritems():
    for key, wedge_dict in time_dict.items():
      
      if( ( wedge_dict['begin'] <= items[1].year) & ( wedge_dict['end'] >= items[1].year ) ):
        range_list.append(key)
        index_s.append(items[0])

    range_list.append( rtr_key)
    index_s.append(items[0])

  sr = pd.Series(range_list)
  sr.index = index_s

  return sr
```

```python
def get_time_section_new(date_row):
  rtr_key = 1200
  single_date = date_row['birthday']
  year = single_date.year
  for key, wedge_dict in time_dict.items():
    
    if( ( wedge_dict['begin'] <= year) & ( wedge_dict['end'] >= year ) ):
      return key

  return rtr_key

  
```

```python
get_time_section(1874)
```

```python
df_period['time_period'] = df_period.apply(axis=1,  func=get_time_section_new )
```
