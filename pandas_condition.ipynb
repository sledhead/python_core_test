```python
df = pd.read_csv('supermarket_sales.csv')
display(df)
```

```python
df_test = df.copy()
```

```python
df.groupby(['Payment','Branch','Customer type'])['Quantity'].count()
```

```python
df_sub_filter = ((df['Payment'] == 'Cash') & (df['Branch'] == 'A') & (df['Customer type'] == 'Member'))
```

```python
df_sub = df[df_sub_filter]
df_sub
```

```python
df_sub.info()
```

```python
df_sub.style.apply(lambda x: ['background:red' if x < 5.5 else 'background:green' for x in df_sub.Rating], axis=0)
```

```python
df_special_format = df_sub.style\
  .applymap(lambda x: 'background-color: %s' % 'red' if x < 5.5 else 'background-color: %s' % 'green', subset=['Rating'])\
  .applymap(lambda x: 'background-color: %s' % 'red' if x < 7 else 'background-color: %s' % 'green', subset=['Quantity'])\
  .highlight_min(subset=['Quantity','Rating'])
```

```python
df_special_format
```

```python
df_special_format.to_excel('test_me.xlsx', engine='openpyxl')
```

```python
df_sub.groupby('Product line')['Rating'].count()
```

```python
df_smaller = df_sub[df_sub['Product line'] == 'Home and lifestyle']
df_smaller
```

```python
df_smaller_style = df_smaller.style\
  .applymap(lambda x: 'background-color: %s' % 'red' if x < 5.5 else 'background-color: %s' % 'green', subset=['Rating'])\
  .applymap(lambda x: 'background-color: %s' % 'red' if x < 7 else 'background-color: %s' % 'green', subset=['Quantity'])\
  .applymap(lambda x: 'background-color: %s' % 'gray' if x == 'Male' else 'background-color: %s' % 'pink', subset=['Gender'])\
  .highlight_min(subset=['Quantity','Rating','Total'])

df_smaller_style
```

```python
df_smaller_style.to_excel('test_me.xlsx', engine='openpyxl')
```
