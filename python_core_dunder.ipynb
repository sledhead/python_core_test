```python

class MontanaPeaks:

  #init function
  def __init__(self, peak_string,height):
    self.peak_string = peak_string
    self.height = height

  def __repr__(self) -> str:
      return 'Object: {}'.format(self.peak_string)

  def __add__(self, other):
    return self.peak_string +  ' ' + other


  def __len__(self):
    return self.height
```

```python

if __name__ == '__main__':

  South_Peak = MontanaPeaks('Boorman',4575)

  print(South_Peak)

  print(South_Peak + 'heaven peak')

  print(f'Give me the len: {len(South_Peak)}')
  
```
